# Combining datasets

```python
import pandas as pd
```

```python
left = pd.DataFrame({'col1': [1,2,3,4,5], 'col2': ['a', 'b', 'c', 'd', 'e']})
right = pd.DataFrame({'col1': [3,4,5,6,7], 'col3': ['f', 'g', 'h', 'i', 'j']})

print(left)
print(right)
```

```python
pd.merge(left=left, right=right, on='col1', how='inner')
```

```python
pd.merge(left, right, on='col1', how='left')
```

```python
pd.merge(left, right, on='col1', how='right')
```

```python
pd.merge(left, right, on='col1', how='outer')
```

```python
df1 = pd.DataFrame({'account_id': [1, 2, 3, 11382], 
                    'gender': ['female', 'male', 'female', 'male'], 
                    'age': [55, 25, 29, 39]})
df1
```

```python
df2 = pd.DataFrame({'account_id': [4, 5, 6, 7], 
                    'gender': ['female', 'male', 'female', 'male'], 
                    'age': [19, 28, 14, 15]})
df2
```

```python
pd.concat([df1, df2], axis=0)
```

```python
pd.concat([df1, df2], ignore_index=True)
```
